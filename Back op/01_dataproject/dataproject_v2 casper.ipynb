## 1.2 Instantaneous
$$
\begin{align*}
\pi_t^{12,\alpha} &= \left( \prod_{k=0}^{11} (1 + \pi_{t-k})^{\kappa(k,\alpha)} \right) - 1, \\[6pt]
\kappa(k,\alpha) &= \frac{(T - k)^{\alpha}}{\sum (T - k)^{\alpha} \, }T
\end{align*}
$$

__More explanation would be great__

__also the plot is very much like chatGPT's version of it... consider changing the asestethics__

```python
# 1) Creating function 
# Define kappa
def kappa(alpha, T=12):
    k = np.arange(T)
    return k, (T - k)**alpha / np.sum((T - np.arange(T))**alpha) * T

# The weights for alpha
alphas = [0, 1, 2, 3]

# 2) Plot kappa with weights 
# a) Creating figure
fig, ax = plt.subplots(figsize=(6, 3))
for a in alphas:
    k, w = kappa(a)
    ax.plot(k, w, marker='o', label=fr'$\alpha={a}$')

# b) Labels and style
ax.set_title(r'Weights $\kappa(k, \alpha)$ for Different $\alpha$ Values')
ax.set_xlabel('k (lag)')
ax.set_ylabel(r'$\kappa(k, \alpha)$')
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# 3) Computing inflation year 
def inflation_year(pi, alpha, T=12):
    _, w_k = kappa(alpha, T)
    w = w_k[::-1]  

    # a) Rolling 
    res = pi.rolling(12).apply(
        lambda x: np.prod((1 + x)**w) - 1,
        raw=True
    )
    return res
```

__great!__

```python
# 4) Plots instantaneous inflation since 2019
# a) Creating figure
fig, ax = plt.subplots(figsize=(6, 3))
for a in alphas:
    pi_year = inflation_year(cpi['pi'], a) 
    mask_2019 = cpi['Time'] >= '2019-01-01'
    time_2019 = cpi.loc[mask_2019, 'Time']
    pi_2019 = pi_year[mask_2019]
    
    ax.plot(time_2019, pi_2019, label=fr'$\alpha={a}$')

# b) Labels and style
ax.set_title("Instantaneous 12-month inflation since 2019")
ax.set_ylabel(r'$\pi_t^{12,\alpha}$')
ax.set_xlabel("Year")
ax.legend()
ax.grid(alpha=0.3)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


The instantaneous 12-month inflation peaked around mid-2022 and returned near zero by early 2023. 
This suggests that the post-pandemic inflation surge ended slightly earlier than indicated by the year-over-year measure.
